Write a python program to display IMDB’s Top rated 100 movies’ data (i.e. Name, IMDB rating, Year of 
release).

```python
#requesting for scraping for 1st page
movies=requests.get("https://www.imdb.com/search/title/?groups=top_100&sort=user_rating,desc")
print(movies)
#requesting for Scraping for 2nd Page
movies_page2=requests.get("https://www.imdb.com/search/title/?groups=top_100&sort=user_rating,desc&start=51&ref_=adv_nxt")
print(movies_page2)
```

```python
#Contents of page1
movies_soup=BeautifulSoup(movies.content)
#contents of Page2
movies_soup2=BeautifulSoup(movies_page2.content)
```

```python
content=movies_soup("div",class_="lister-item-content")
print(content)
content2=movies_soup2("div",class_="lister-item-content")
print(content2)
```

```python
#To get the name, year and rating of page1 (First 50 movies) 
name_series=[]
year_series=[]
rating_series=[]
for i in content:
    #Name of the movies
    for j in i.find_all("h3",class_="lister-item-header"):
        for z in j.find_all("a"):
            name_series.append(z.text.replace("\n",""))
    #year of the movies
    for j in i.find_all(class_="lister-item-year text-muted unbold"):
        year_series.append(j.text)
    #Ratings of the movies
    for j in i.find_all("div",class_="inline-block ratings-imdb-rating"):
        rating_series.append(j.text.replace("\n",""))
    
```

```python
#To get the name, year and rating of page2(Last 50 movies)
name_series2=[]
year_series2=[]
rating_series2=[]
for i in content2:
    #Name of the movies
    for j in i.find_all("h3",class_="lister-item-header"):
        for z in j.find_all("a"):
            name_series2.append(z.text.replace("\n",""))
    #year of the movies
    for j in i.find_all(class_="lister-item-year text-muted unbold"):
        year_series2.append(j.text)
    #Ratings of the movies
    for j in i.find_all("div",class_="inline-block ratings-imdb-rating"):
        rating_series2.append(j.text.replace("\n",""))
    
```

```python
#Providing the complete data in a table 
name=name_series+name_series2
year=year_series+year_series2
rating=rating_series+rating_series2
data={"Movie Name":name,"Year Of Release":year,"Ratings":rating}
df=pd.DataFrame(data, columns=['Movie Name',"Year Of Release","Ratings"])
df
```

Write a python program to display IMDB’s Top rated 100 Indian movies’ data (i.e. Name, IMDB rating, Year 
of release).

```python
def Indian_movies():
    #requesting for scraping
    movies=requests.get("https://www.imdb.com/list/ls009997493/")
    print(movies)
    #Contents
    movies_soup=BeautifulSoup(movies.content)
    content=movies_soup("div",class_="lister-item-content")
    #To get the name, year and rating of Indian Movies
    name_series=[]
    year_series=[]
    rat_series=[]
    for i in content:
    #Name of the movies
        for j in i.find_all("h3",class_="lister-item-header"):
            for z in j.find_all("a"):
                name_series.append(z.text.replace("\n",""))
    #year of the movies
        for j in i.find_all(class_="lister-item-year text-muted unbold"):
            year_series.append(j.text)
    #Ratings of the movies

        for j in i.find_all("div",class_="ipl-rating-star small"):
            for z in j.find_all("span",class_="ipl-rating-star__rating"):
                rat_series.append(z.text)
    data={"Movie Name":name_series,"Year Of Release":year_series,"Ratings":rat_series}
    df_indian=pd.DataFrame(data, columns=['Movie Name',"Year Of Release","Ratings"])
    return (df_indian)

#Function calling:
Indian_movies()
```

Write a python program to scrape fresher job listings from ‘https://internshala.com/’. It should include job title, 
company name, CTC, and apply date

```python
job=requests.get("https://internshala.com/fresher-jobs")
job
```

```python
job_soup=BeautifulSoup(job.content)
```

```python
content=job_soup("div",id="internship_list_container_1")
content
```

```python
title=[]
name=[]
ctc=[]
date=[]

for i in content:
    for j in i.find_all("div",class_="heading_4_5 profile"):
        title.append(j.text)
    for j in i.find_all("div",class_="heading_6 company_name"):
        for z in j.find_all("a"):
            name.append(z.text.replace("\n",""))
title
name
```

```python
cc_data={"Job Title":title,"Company_name":name}
cc_df=pd.DataFrame(cc_data,columns=["Job Title","Company_name"])
cc_df
```
